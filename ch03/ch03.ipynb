## 3.1-3.2 Reach, crush, ledger (Listings 3.1-3.2)

```python
import numpy as np

C = np.array([[1, 0, -1], [-1, 1, 0], [0, -1, 1]])
print('C @ (3,3,3):', C @ np.array([3, 3, 3]))
A3 = np.array([[1, 0, 0], [-1, 1, 0], [0, -1, 1]])
for name, M in [('A3', A3), ('C ', C)]:
    r = np.linalg.matrix_rank(M)
    print(f'{name}: rank {r}, nullity {M.shape[1] - r}')
```

## 3.4-3.6 Elimination, LU, the machine verified (Listings 3.3-3.4)

```python
A = np.array([[1., 2, 1], [2, 5, 4], [1, 3, 5]])
L = np.array([[1., 0, 0], [2, 1, 0], [1, 1, 1]])
U = np.array([[1., 2, 1], [0, 1, 2], [0, 0, 2]])
print('max |L @ U - A|:', np.abs(L @ U - A).max())
```

```python
b = np.array([5., 13, 12])
x_hat = np.linalg.solve(A, b)
print('x_hat   :', x_hat)
print('residual:', np.abs(b - A @ x_hat).max())
```

## 3.7 The door (Listing 3.5, Figure 3.2)

```python
import pandas as pd
import matplotlib.pyplot as plt

zoning  = pd.read_csv('data/zoning.csv')
listing = pd.read_csv('data/listing.csv')
sale    = pd.read_csv('data/sale.csv')
housing = pd.merge(zoning, listing, on='Id')
housing = pd.merge(housing, sale, on='Id').set_index('Id')
X = housing[['GrLivArea', 'OverallQual']].to_numpy(float)
y = housing['SalePrice'].to_numpy(float)
A2 = X[:2]; b2 = y[:2]
w = np.linalg.solve(A2, b2)
print('w       :', np.round(w, 2))
print('residual:', np.abs(b2 - A2 @ w).max())
print('house 3 pred:', round(X[2] @ w), 'actual', int(y[2]))
```

```python
w_best, *_ = np.linalg.lstsq(X, y, rcond=None)
print('lstsq w:', np.round(w_best, 2))
pred = X @ w_best
plt.figure(figsize=(7, 5.5))
plt.scatter(X[:, 0], y, s=8, alpha=0.3, label='actual')
plt.scatter(X[:, 0], pred, s=8, alpha=0.3, label='predicted')
plt.xlabel('GrLivArea (sq ft)'); plt.ylabel('SalePrice ($)')
plt.title('The best-fit band misses everything a little, on purpose')
plt.grid(True, alpha=0.3); plt.legend()
plt.savefig('figures/fig_price_vs_sqft.png', dpi=150, bbox_inches='tight')
plt.show()
```

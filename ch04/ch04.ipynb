## 4.0-4.2 Eigenpairs verified + diagonalization (Listing 4.1)

```python
import numpy as np

A = np.array([[2., 1], [1, 2]])
print('A@(1,1) :', A @ [1, 1])
print('A@(1,-1):', A @ [1, -1])
X = np.array([[1., 1], [1, -1]])
Lam = np.diag([3., 1])
Xinv = np.linalg.inv(X)
print('max |X Lam Xinv - A|  :', np.abs(X @ Lam @ Xinv - A).max())
A5_eig = X @ np.diag([3.**5, 1.]) @ Xinv
print('max |A^5 (eig) - A^5| :', np.abs(A5_eig - np.linalg.matrix_power(A, 5)).max())
```

## 4.3 The sines, confronted (Listings 4.2-4.3, Figure 4.2)

```python
n = 200
x = np.linspace(0, 2*np.pi, n)
h = x[1] - x[0]
K = (np.eye(n, k=1) - 2*np.eye(n) + np.eye(n, k=-1)) / h**2
vals, vecs = np.linalg.eigh(-K * h * h)
order = np.argsort(vals)
vals, vecs = vals[order], vecs[:, order]
j = np.arange(1, n + 1)
for k in (1, 3):
    formula = 2 - 2*np.cos(k*np.pi/(n + 1))
    sine = np.sin(j*k*np.pi/(n + 1)); sine /= np.linalg.norm(sine)
    agree = abs(sine @ vecs[:, k - 1])
    print(f'k={k}: lambda {vals[k-1]:.10f} vs formula {formula:.10f},'
          f' sine agreement {agree:.10f}')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7.5, 4.5))
for k in range(3):
    plt.plot(vecs[:, k], label=f'eigenvector {k + 1}')
plt.legend(); plt.title('The first eigenvectors of K')
plt.savefig('figures/fig_k_eigenvectors.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4.4 The power method (Listing 4.4)

```python
v = np.array([1., 0.])
for _ in range(20):
    w = A @ v
    v = w / np.linalg.norm(w)
lam = v @ A @ v
print('direction:', np.round(v, 6))
print('stretch  :', round(lam, 8))
```

```python
# Listing 4.5 (the convergence, drawn) -> Figure 4.3
v = np.array([1., 0.])
gap = []
for _ in range(12):
    w = A @ v
    v = w / np.linalg.norm(w)
    gap.append(3 - v @ A @ v)
plt.semilogy(np.arange(1, 13), gap, 'o-')
plt.xlabel('iteration')
plt.ylabel('gap to the true stretch')
plt.savefig('figures/fig_power_convergence.png', dpi=150,
            bbox_inches='tight')
plt.show()
```

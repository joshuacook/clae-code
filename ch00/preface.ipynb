# Preface — the difference-matrix exhibit

Reproduces the preface's one exhibit: a derivative, on a grid, is a matrix.

Run: `clae-py -m nbconvert --to notebook --execute --inplace preface.ipynb`

```python
import numpy as np

x = np.linspace(0, 2*np.pi, 1000)
h = x[1] - x[0]
D = (np.eye(1000, k=1) - np.eye(1000)) / h    # -1 and 1 down the diagonals

np.abs(D @ np.sin(x) - np.cos(x))[:-1].max()  # forward difference
```

```python
# Listing P.2 (the derivative, drawn) -> Figure P.3
import os
import matplotlib.pyplot as plt

os.makedirs('figures', exist_ok=True)
plt.plot(x, np.sin(x), label='input: sin(x)')
plt.plot(x[:-1], (D @ np.sin(x))[:-1], label='output: D @ sin(x)')
plt.plot(x, np.cos(x), 'k--', lw=1, label='cos(x)')
plt.legend()
plt.savefig('figures/fig_preface_derivative.png', dpi=150,
            bbox_inches='tight')
plt.show()
```
